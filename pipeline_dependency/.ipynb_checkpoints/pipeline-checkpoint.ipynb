```python
def dependency_pipeline(node, initial=[], l=[]):

    dependencies = get_dependencies(node)
    missing_dependencies = (set(dependencies)-set(l)) -set(initial)
    
    # Base case descendants == initial or descendants == None 
    if len(missing_dependencies)==0:
        return initial
    
    
    # Recursive function 
    else:
            
        for val in missing_dependencies:

            if val not in l:
                print('Executing {}'.format(val))
                ls = dependency_pipeline(val,initial, l)
                [ls.remove(item) for item in l if item in ls]
                # ls = list(set(ls)-set(l))
                print('Appending {}'.format(ls))
                l = l + ls
                l.append(val)

        return l

```

```python
def get_dependencies(node):
    return list(relations[relations['dest']==node]['scr'])
```

```python
import networkx as nx
```

```python
G = nx.DiGraph()

G.add_edges_from(list(zip(relations['scr'],relations['dest'])))

plt.figure(figsize=(15,12))
nx.draw(G, with_labels = True, node_size=1500, alpha=0.8)
plt.draw()
```

```python
print("Downstream Edges of '36' (just example)-->")
print(list(nx.edge_dfs(G,36, orientation='reverse')))
```
